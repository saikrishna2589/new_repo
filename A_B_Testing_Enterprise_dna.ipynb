```python
#scipy is one of the best libraries for statistical analysis
from scipy import stats
ad_group=df[df['test group']=='ad']['converted']

psa_group=df[df['test group']=='psa']['converted']
```

```python
ad_group
```

```python
# t-test ,which is test for independence
t_stat,p_val=stats.ttest_ind(ad_group,psa_group,equal_var=False)
```

```python
# p-value shpws print(f"The tstat is {t_stat} and the p_value is {p_val} ")
```

T-Statistic: The t-statistic of 8.657 indicates a substantial difference between the groups. It suggests that there is a significant difference between group A and group B.

P-Value: The p-value is very low (5.107607598758945e-18), essentially zero when rounded. This indicates that the probability of observing such extreme results (or more extreme) under the null hypothesis (i.e., no difference between groups) is extremely unlikely.

Given the low p-value, we can reject the null hypothesis and conclude that there is a statistically significant difference between group A and group B.

In summary, the A/B test result suggests that there is a significant difference between the two groups, with group B likely performing differently from group A.
